```python
#function to get number of bounding boxes
def getBoxes():
    bb = []
    lab = []
    for i in range(0, 4):
        bb.append([0.0, 0.0, 0.0, 0.0])
        lab.append(0)
    return np.asarray(bb), np.asarray(lab)    
```

```python
#splitting dataset into train and test where application using 80% images for training and 20% for testing
split = train_test_split(X, Y, boxes, test_size=0.20, random_state=42)
(trainImages, testImages) = split[:2]
(trainLabels, testLabels) = split[2:4]
(trainBBoxes, testBBoxes) = split[4:6]
print("Train & Test Data Size")
print("80% images used for training : "+str(trainImages.shape[0]))
print("20% images used for testing : "+str(testImages.shape[0]))
```

```python
#function to calculate model accuracy and other metrics
def calculateMetrics(algorithm, predict, y_test):
    labels = ['0', '1', '2', '3']
    a = accuracy_score(y_test,predict)*100
    p = precision_score(y_test, predict,average='macro') * 100
    r = recall_score(y_test, predict,average='macro') * 100
    f = f1_score(y_test, predict,average='macro') * 100
    print(algorithm+" Accuracy  :  "+str(a))
    print(algorithm+" Precision : "+str(p))
    print(algorithm+" Recall    : "+str(r))
    print(algorithm+" FScore    : "+str(f))    
    conf_matrix = confusion_matrix(y_test, predict) 
    plt.figure(figsize =(6, 4)) 
    ax = sns.heatmap(conf_matrix, xticklabels = labels, yticklabels = labels, annot = True, cmap="viridis" ,fmt ="g");
    ax.set_ylim([0,len(labels)])
    plt.title(algorithm+" Confusion matrix") 
    plt.xticks(rotation=90)
    plt.ylabel('True class') 
    plt.xlabel('Predicted class') 
    plt.show()   
```

```python
def detectVehicle(img_path):
    labels = ["0", "1", "2", "3"]
    img = cv2.imread(img_path)
    img = cv2.resize(img, (300, 300))
    data = []
    data.append(img)
    data = np.asarray(data)
    data = data.astype('float32')
    data = data/255
    predicts = rtsnet_model.predict(data)
    bounding_boxes = predicts[0]
    predicted_labels = predicts[1]
    predicted_labels = predicted_labels.ravel()
    bounding_boxes = bounding_boxes.ravel()
    predicted_labels = np.where(predicted_labels > 0.1, 1, 0)
    img = cv2.imread(img_path)
    img = cv2.resize(img, (300, 300))
    i = 0
    index = 0
    while i < (len(bounding_boxes)):
        xmin = bounding_boxes[i]
        i += 1
        ymin = bounding_boxes[i]
        i += 1
        xmax = bounding_boxes[i]
        i += 1
        ymax = bounding_boxes[i]
        i += 1
        if xmin > 0 and ymin > 0 and xmax > 0 and ymax > 0:
            xmin, ymin, xmax, ymax = pbx.convert_bbox((xmin, ymin, xmax, ymax), from_type="yolo", to_type="voc", image_size=(300, 300))
            cv2.rectangle(img, (xmin, ymin), (xmax, ymax), (255, 0, 0), 2)
            if predicted_labels[index] == 1:
                cv2.putText(img, labels[index], (xmin, ymin), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0,255,0),thickness=2)
        index += 1 
    plt.figure(figsize=(4,3))    
    plt.imshow(img)
    plt.show()
```
